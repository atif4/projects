```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
file = 'https://aegis4048.github.io/downloads/notebooks/sample_data/unconv_MV_v5.csv'
df = pd.read_csv(file)
df.head()
```

```python
def get_plt_scatter(col_id1, col_id2):
    plt.scatter(df[col_id1],df[col_id2],color='red',label='scatter')
    plt.xlabel(col_id1)
    plt.ylabel(col_id2)
    plt.title(col_id1+' V/S '+col_id2)
    plt.show()
get_plt_scatter('Por', 'Prod')
```

```python
df['Por'].shape
```

```python
X = df['Por'].values.reshape(-1,1) # it used to convert the dimension 
y = df['Prod'].values
```

```python
X.shape
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
pred = lr.predict(X)
```

```python
r2 = lr.score(X,y)
```

```python
plt.style.use('default')
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(X,pred, color='k', label='Regression model')
ax.scatter(X, y, edgecolor='k', facecolor='red', alpha=0.7, label='Sample data')
ax.scatter(X, pred, edgecolor='k', facecolor='blue', alpha=0.7, label='r2')
ax.set_ylabel('Prod', fontsize=10)
ax.set_xlabel('Por', fontsize=10)
ax.text(0.3, 0.1, 'aegis4048.github.io', fontsize=3, ha='center', va='center',transform=ax.transAxes, color='red', alpha=0.2)
ax.legend(facecolor='white', fontsize=8)
ax.set_title('$R^2= %.2f$' % r2, fontsize=10)
fig.tight_layout()
```
